## Basic Theory of Electromagnetic Waves

Electromagnetic (EM) waves are disturbances that propagate through space or a medium, carrying energy. They are composed of oscillating electric and magnetic fields that are perpendicular to each other and to the direction of propagation.

Key characteristics of EM waves include:

*   **Wavelength ($\lambda$)**: The distance between two consecutive crests or troughs of the wave.
*   **Frequency ($f$)**: The number of oscillations per unit time.
*   **Speed of light ($c$)**: The speed at which EM waves travel in a vacuum (approximately $3 \times 10^8$ m/s).

These properties are related by the equation: $c = \lambda f$.

In the context of light, which is a form of electromagnetic radiation, the electric field plays a crucial role in interactions with matter, particularly in techniques like Optical Coherence Tomography (OCT). The oscillating electric field exerts a force on charged particles, leading to phenomena like reflection, refraction, and scattering, which are fundamental to how OCT works.

## ทฤษฎีพื้นฐานของคลื่นแม่เหล็กไฟฟ้า

คลื่นแม่เหล็กไฟฟ้า (EM) คือการรบกวนที่แพร่กระจายผ่านอวกาศหรือตัวกลาง โดยนำพาพลังงานไปด้วย คลื่นเหล่านี้ประกอบด้วยสนามไฟฟ้าและสนามแม่เหล็กที่สั่นตั้งฉากกันและตั้งฉากกับทิศทางการแพร่กระจาย

ลักษณะสำคัญของคลื่นแม่เหล็กไฟฟ้า ได้แก่:

*   **ความยาวคลื่น ($\lambda$)**: ระยะห่างระหว่างยอดคลื่นหรือท้องคลื่นสองลูกที่ติดกัน
*   **ความถี่ ($f$)**: จำนวนรอบของการสั่นต่อหน่วยเวลา
*   **ความเร็วแสง ($c$)**: ความเร็วที่คลื่นแม่เหล็กไฟฟ้าเดินทางในสุญญากาศ (ประมาณ $3 \times 10^8$ เมตรต่อวินาที)

คุณสมบัติเหล่านี้มีความสัมพันธ์กันตามสมการ: $c = \lambda f$

ในบริบทของแสง ซึ่งเป็นรูปแบบหนึ่งของรังสีแม่เหล็กไฟฟ้า สนามไฟฟ้ามีบทบาทสำคัญในการโต้ตอบกับสสาร โดยเฉพาะอย่างยิ่งในเทคนิคต่างๆ เช่น Optical Coherence Tomography (OCT) สนามไฟฟ้าที่สั่นจะออกแรงกระทำต่ออนุภาคที่มีประจุ นำไปสู่ปรากฏการณ์ต่างๆ เช่น การสะท้อน การหักเห และการกระเจิง ซึ่งเป็นพื้นฐานสำคัญในการทำงานของ OCT

## Define parameters

### Subtask:
Define the wavelength, amplitude, and other relevant parameters for the electric field.


**Reasoning**:
Define the variables and calculate the angular frequency and wave number as instructed.



```python
import numpy as np

# Define parameters
wavelength = 500e-9  # Wavelength in meters (500 nm)
amplitude = 1.0  # Amplitude of the electric field
c = 3e8  # Speed of light in meters per second

# Calculate angular frequency (omega) and wave number (k)
omega = 2 * np.pi * c / wavelength
k = 2 * np.pi / wavelength

print(f"Wavelength (λ): {wavelength:.2e} m")
print(f"Amplitude: {amplitude}")
print(f"Speed of light (c): {c:.0e} m/s")
print(f"Angular frequency (ω): {omega:.2e} rad/s")
print(f"Wave number (k): {k:.2e} rad/m")
```

## Generate Time and Space Arrays

### Subtask:
Create arrays representing time and space to calculate the electric field over a range.

```python
# Generate time and space arrays
# Define the range of time and space for visualization
time_range = np.linspace(0, 2 * wavelength / c, 1000)  # Two periods of oscillation
space_range = np.linspace(0, 2 * wavelength, 1000)  # Two wavelengths

print("Time array shape:", time_range.shape)
print("Space array shape:", space_range.shape)
```

## Electric Field Construction

The electric field of a simple plane electromagnetic wave, like the one we are simulating, can be described by a sinusoidal function. This function depends on both space and time, and its amplitude represents the strength of the electric field at a given point and time.

In our case, we assumed the wave is propagating in the z-direction and the electric field is polarized in the x-direction. The electric field component ($E_x$) at a specific point in space ($z$) and time ($t$) is calculated using the formula:

$E_x(z, t) = E_0 \cos(kz - \omega t)$

where:
*   $E_0$ is the amplitude of the electric field.
*   $k$ is the wave number, related to the spatial frequency ($\lambda$).
*   $\omega$ is the angular frequency, related to the temporal frequency ($f$).
*   $z$ is the position in the direction of propagation.
*   $t$ is the time.

We calculated the electric field in space at a fixed time ($t=0$) and the electric field in time at a fixed position ($z=0$) to visualize its behavior along these dimensions.

## การสร้างสนามไฟฟ้า

สนามไฟฟ้าของคลื่นแม่เหล็กไฟฟ้าแบบระนาบอย่างง่าย เช่น คลื่นที่เรากำลังจำลอง สามารถอธิบายได้ด้วยฟังก์ชันไซน์ ฟังก์ชันนี้ขึ้นอยู่กับทั้งอวกาศและเวลา และแอมพลิจูดของมันแสดงถึงความแรงของสนามไฟฟ้า ณ จุดและเวลาที่กำหนด

ในกรณีของเรา เราสมมติว่าคลื่นกำลังแพร่กระจายไปในทิศทางแกน z และสนามไฟฟ้ามีโพลาไรซ์ในทิศทางแกน x ส่วนประกอบของสนามไฟฟ้า ($E_x$) ณ จุดที่กำหนดในอวกาศ ($z$) และเวลา ($t$) คำนวณได้โดยใช้สูตร:

$E_x(z, t) = E_0 \cos(kz - \omega t)$

โดยที่:
*   $E_0$ คือแอมพลิจูดของสนามไฟฟ้า
*   $k$ คือเลขคลื่น ซึ่งสัมพันธ์กับความถี่เชิงพื้นที่ ($\lambda$)
*   $\omega$ คือความถี่เชิงมุม ซึ่งสัมพันธ์กับความถี่เชิงเวลา ($f$)
*   $z$ คือตำแหน่งในทิศทางการแพร่กระจาย
*   $t$ คือเวลา

เราคำนวณสนามไฟฟ้าในอวกาศที่เวลาคงที่ ($t=0$) และสนามไฟฟ้าในเวลาที่ตำแหน่งคงที่ ($z=0$) เพื่อแสดงภาพพฤติกรรมของมันตามมิติเหล่านี้

## Calculate Electric Field Components

### Subtask:
Calculate the electric field components (e.g., Ex, Ey) based on the defined parameters and generated arrays.

```python
# Calculate electric field components
# For a simple plane wave propagating in the z-direction with electric field polarized in the x-direction
Ex = amplitude * np.cos(k * space_range - omega * time_range[0]) # Electric field at t=0 along space
Et = amplitude * np.cos(-omega * time_range + k * space_range[0]) # Electric field at z=0 along time

print("Electric field (Ex) shape:", Ex.shape)
print("Electric field (Et) shape:", Et.shape)
```

## Visualize Electric Field

### Subtask:
Plot the electric field components to demonstrate the wave's behavior.

```python
import matplotlib.pyplot as plt

# Visualize the electric field in space at a fixed time (t=0)
plt.figure(figsize=(10, 4))
plt.plot(space_range * 1e9, Ex)  # Convert space to nanometers for better visualization
plt.xlabel("Position (nm)")
plt.ylabel("Electric Field Amplitude")
plt.title("Electric Field in Space (at t=0)")
plt.grid(True)
plt.show()

# Visualize the electric field in time at a fixed position (z=0)
plt.figure(figsize=(10, 4))
plt.plot(time_range * 1e15, Et)  # Convert time to femtoseconds for better visualization
plt.xlabel("Time (fs)")
plt.ylabel("Electric Field Amplitude")
plt.title("Electric Field in Time (at z=0)")
plt.grid(True)
plt.show()
```

## Theory of Wave Intensity

The intensity of a wave is a measure of the power carried by the wave per unit area. For electromagnetic waves, the intensity ($I$) is proportional to the square of the electric field amplitude ($E$) and the square of the magnetic field amplitude ($B$).

The relationship can be expressed as:

$I \propto E^2 \propto B^2$

More precisely, the intensity of an electromagnetic wave is given by the magnitude of the Poynting vector, averaged over one period of the wave. In a vacuum, this can be simplified to:

$I = \frac{1}{2} c \epsilon_0 E_0^2$

where:
*   $c$ is the speed of light in vacuum
*   $\epsilon_0$ is the permittivity of free space
*   $E_0$ is the electric field amplitude

In the context of the electric field calculation we performed, squaring the electric field components ($Ex^2$ and $Et^2$) gives us a value proportional to the intensity at each point in space and time. While not the absolute intensity, these squared values accurately represent the relative intensity variations of the wave.

## ทฤษฎีความเข้มคลื่น

ความเข้มของคลื่นคือการวัดกำลังที่คลื่นนำพาต่อหน่วยพื้นที่ สำหรับคลื่นแม่เหล็กไฟฟ้า ความเข้ม ($I$) เป็นสัดส่วนกับกำลังสองของแอมพลิจูดสนามไฟฟ้า ($E$) และกำลังสองของแอมพลิจูดสนามแม่เหล็ก ($B$)

ความสัมพันธ์สามารถแสดงได้ดังนี้:

$I \propto E^2 \propto B^2$

ที่แม่นยำกว่านั้น ความเข้มของคลื่นแม่เหล็กไฟฟ้าจะเท่ากับขนาดของเวกเตอร์ Poynting โดยเฉลี่ยในช่วงเวลาหนึ่งของคลื่น ในสุญญากาศ สามารถทำให้ง่ายขึ้นได้เป็น:

$I = \frac{1}{2} c \epsilon_0 E_0^2$

โดยที่:
*   $c$ คือความเร็วแสงในสุญญากาศ
*   $\epsilon_0$ คือค่าสภาพยอมของสุญญากาศ
*   $E_0$ คือแอมพลิจูดสนามไฟฟ้า

ในบริบทของการคำนวณสนามไฟฟ้าที่เราทำ กำลังสองของส่วนประกอบสนามไฟฟ้า ($Ex^2$ และ $Et^2$) จะให้ค่าที่เป็นสัดส่วนกับความเข้ม ณ แต่ละจุดในอวกาศและเวลา แม้ว่าจะไม่ใช่ความเข้มสัมบูรณ์ แต่ค่ากำลังสองเหล่านี้แสดงถึงการเปลี่ยนแปลงความเข้มสัมพัทธ์ของคลื่นได้อย่างถูกต้อง

## Calculate wave intensity

### Subtask:
Calculate the intensity of the electric field based on the previously calculated components.


**Reasoning**:
Calculate the intensity of the electric field along space and time by squaring the respective electric field components.



```python
# Calculate the intensity of the electric field
Intensity_x = Ex**2
Intensity_t = Et**2

print("Intensity in space shape:", Intensity_x.shape)
print("Intensity in time shape:", Intensity_t.shape)
```

## Visualize wave intensity

### Subtask:
Plot the calculated wave intensity as a function of space and/or time.


**Reasoning**:
Plot the calculated wave intensity as a function of space and time based on the instructions.



```python
# Plot the intensity in space at a fixed time (t=0)
plt.figure(figsize=(10, 4))
plt.plot(space_range * 1e9, Intensity_x)  # Convert space to nanometers for better visualization
plt.xlabel("Position (nm)")
plt.ylabel("Intensity")
plt.title("Electric Field Intensity in Space (at t=0)")
plt.grid(True)
plt.show()

# Plot the intensity in time at a fixed position (z=0)
plt.figure(figsize=(10, 4))
plt.plot(time_range * 1e15, Intensity_t)  # Convert time to femtoseconds for better visualization
plt.xlabel("Time (fs)")
plt.ylabel("Intensity")
plt.title("Electric Field Intensity in Time (at z=0)")
plt.grid(True)
plt.show()
```
